```python
y=data["Performance_Score"]
X=data[["Years_At_Company","Monthly_Salary","Overtime_Hours","Promotions","Employee_Satisfaction_Score"]]
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
#Scaled X train
X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
#Scaled X test
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
import joblib
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
joblib.dump(model, "model.pkl")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(predictions):
    print("Accuracy score on model is{}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors": [3,5,7,9,11],
              "weights":["uniform","distance"]}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
```

```python
gridkn.fit(X_train, y_train)
```

```python
gridkn.best_params_
```

```python
predictions = gridkn.predict(X_test)
```

```python
modelperformance(predictions)
```

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
param_grid = {
    "C" : [0.01, 0.1, 0.5, 1],
    "kernel" : ["linear","rbf","poly"]
}
```

```python
gridsvc = GridSearchCV(svm, param_grid, cv=5)
```

```python
gridsvc.fit(X_train,y_train)
```

```python
y_pred = gridsvc.predict(X_test)
```

```python
modelperformance(y_pred)
```

```python
joblib.dump(gridsvc,"model.pkl")
```
